Seguindo o exemplo do algoritmo de binarização apresentado em nossa última aula, realize a implementação em Python para transformar uma imagem colorida para níveis de cinza (0 a 255) e para binarizada (0 e 255), preto e branco.

```python
import cv2
from matplotlib import pyplot as plt

# Configurar o matplotlib para mostrar as imagens inline no notebook jupyter
%matplotlib inline
```

## Realizando a transformação

```python
# Carregando a imagem original
imagem = cv2.imread('deadpool-wolverine.jpg')

# Convertendo a imagem para tons de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Nova figura para mostrar as imagens
plt.figure(figsize=(10, 5))

# Mostrando a imagem original
plt.subplot(1, 2, 1)  # 1 linha, 2 colunas, 1ª posição
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.title('Imagem Colorida')
plt.axis('off')  # Ocultar os eixos

# Mostrando a imagem em tons de cinza
plt.subplot(1, 2, 2)  # 1 linha, 2 colunas, 2ª posição
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Tons de Cinza')
plt.axis('off')  # Ocultar os eixos

# Ajustandolayout para evitar sobreposição (dos eixos)
plt.tight_layout()  # Para não printar os eixos

plt.show()
```

```python
# Aplicando um limiar para converter para preto e branco (estando em tons de cinza já)
_, imagem_pb = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)

# Nova figura para mostrar as imagens
plt.figure(figsize=(10, 5))


# Mostrando a imagem em tons de cinza
plt.subplot(1, 2, 1)  # 1 linha, 2 colunas, 1ª posição
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Tons de Cinza')
plt.axis('off')  # Ocultar os eixos

# Mostrar a imagem em preto e branco
plt.subplot(1, 2, 2)  # 1 linha, 2 colunas, 2ª posição
plt.imshow(imagem_pb, cmap='gray')
plt.title('Imagem em Preto e Branco')
plt.axis('off')  # Ocultar os eixos

# Ajustandolayout para evitar sobreposição (dos eixos)
plt.tight_layout()  # Para não printar os eixos

plt.show()
```
